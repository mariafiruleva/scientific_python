### 2.1
First create iterator via my_iter = map(lambda ...) construction that takes [1,2,3,4,5,6]
and returns True if there is no remainder after division of the lists element on 3 and
returns False otherwise

```python
my_iter = map(lambda x: False if x % 3 else True, range(1, 7))
```

### 2.2 
In infinite loop proceed the iterator using the next() function on it. Observe the appearing
error in this loop.

```python
while True:
    print(next(my_iter))
```

### 2.3 
Write except block catching this particular error

```python
my_iter = map(lambda x: False if x % 3 else True, range(1, 7))
while True:
    try:
        print(next(my_iter))
    except StopIteration:
        print("You've already iterated through all elements from my_iter iterator")
        break
```

What if I don't reassign my_iter iterator?

```python
while True:
    try:
        print(next(my_iter))
    except StopIteration:
        print("You've already iterated through all elements from my_iter iterator")
        break
```

We can itarate over the iterator only one times.

```python
my_iter = map(lambda x: False if x % 3 else True, range(1, 7))
print(list(my_iter))
print(list(my_iter)) # empty list
```

## 2.4 Based on this usecase try to understand in which cases iterators might be more preferable than lists.

Iterators more preferable than lists if:

1) You have a list with a lot of elements, iterators more preferable since iterator takes less memory than list

2) Our aim is to iterate over the massive only ones, we can use iterator. If you try to iterate again, the program will immediately give a StopIteration error (in list case, program will iterate over the list elements again).
